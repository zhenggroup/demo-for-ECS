# Plotting example
## Demo for ECS Webinar

```python
import pandas as pd
import matplotlib.pyplot as plt

# Variables and basic operations
# Variables: x and y lists store the data points
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a DataFrame
data = {'X': x,'Y': y}
print(data)

# Pandas DataFrame: Organizes the data for easy manipulation and plotting.
df = pd.DataFrame(data)
print(df)

# Plotting
# Matplotlib Plotting: Creates a line plot to visualize the relationship between x and y.
fig, axs = plt.subplots(1,1,figsize=(5,5),layout='constrained')
axs.plot(df['X'], df['Y'], marker='o', linestyle='-')

axs.set_title('Line Plot of X vs Y', fontsize = 10)
axs.set_xlabel('X')
axs.set_ylabel('Y')
plt.grid(False)
plt.savefig('ECS webinar example.pdf')
plt.show()
```
