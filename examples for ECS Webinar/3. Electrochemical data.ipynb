# Electrochemical data processing and plotting
## Demo for ECS Webinar

```python
import pandas as pd
import matplotlib.pyplot as plt

CVdata = pd.read_excel('example data 1.xlsx')
print(CVdata)
```

```python
CVdata = pd.read_excel('example data 1.xlsx', skiprows=5)
print(CVdata)
```

```python
new_labels = ['Potential (V)', 'Current (µA)']
CVdata = pd.read_excel('example data 1.xlsx', skiprows=5, names=new_labels)
print(CVdata)
```

```python
fig, axs = plt.subplots(1,1,figsize=(4,3),layout='constrained')
axs.plot(CVdata['Potential (V)'],CVdata['Current (µA)'])
axs.set_xlabel('Potential / V')
axs.set_ylabel('Current / µA')
axs.set_xlim(0.2,0.72)
```

```python
ECSA = 0.75 #in cm2

fig, axs = plt.subplots(1,1,figsize=(4,3),layout='constrained')
axs.plot(CVdata['Potential (V)'],CVdata['Current (µA)']/ECSA/1000)
axs.set_xlabel('Potential / V')
axs.set_ylabel(r'Current density / mA $cm^{-2}$')
axs.set_xlim(0.2,0.72)
```
